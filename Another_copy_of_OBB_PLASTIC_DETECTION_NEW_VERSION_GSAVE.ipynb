```python
import os

# Check the actual structure of the downloaded dataset
dataset_base_path = '/content/drive/MyDrive/waste_sorting_training/dataset'
print("Contents of dataset folder:")
for item in os.listdir(dataset_base_path):
    item_path = os.path.join(dataset_base_path, item)
    if os.path.isdir(item_path):
        print(f"📁 {item}/")
        # List subdirectories
        for subitem in os.listdir(item_path):
            subitem_path = os.path.join(item_path, subitem)
            if os.path.isdir(subitem_path):
                print(f"   📁 {subitem}/")
                # List files in subdirectories
                for file in os.listdir(subitem_path)[:5]:  # Show first 5 files
                    print(f"      📄 {file}")
                if len(os.listdir(subitem_path)) > 5:
                    print(f"      ... and {len(os.listdir(subitem_path))-5} more files")
            else:
                print(f"   📄 {subitem}")
    else:
        print(f"📄 {item}")
```
